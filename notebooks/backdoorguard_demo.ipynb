# BackdoorGuard Demo

This notebook runs the MVP auditor on a curated demo dataset and prints structured results.

```
from pathlib import Path
import sys
sys.path.append(str(Path('../src').resolve()))

from backdoorguard.auditor import BackdoorGuardAuditor
from backdoorguard.io_utils import load_json
import json
```

```
auditor = BackdoorGuardAuditor(model_path='../models/baseline.joblib')
samples = load_json('../data/demo_logs.json')
len(samples)
```

```
results = [auditor.audit(sample).to_dict() for sample in samples]
results[:2]
```

```
print(json.dumps(results, indent=2, ensure_ascii=False))
```
